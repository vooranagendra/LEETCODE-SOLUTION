<B>SUBSETS

```python
class Solution(object):
    def subsets(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        subsets=[]
        bins=[]
        for i in range(0,2**(len(nums))):
            curr_val=[]
            while i>0:
                if i%2==0:
                    curr_val.insert(0,0)
                else:
                    i-=1
                    curr_val.insert(0,1)
                i/=2
            if len(curr_val)==0:
                bins.append([0])
            else:
                bins.append(curr_val)
        for i in bins:
            curr_subset=[]
            for j in range(1,len(i)+1):
                if i[-j]==1:
                    curr_subset.insert(0,nums[-j])
            subsets.append(curr_subset)
        return subsets
```

<b>single number 2

```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        bin_vals=[]
        for i in nums:
            curr_val=""
            while i>0:
                if i%2==0:
                    temp="0"+curr_val
                    curr_val=temp
                else:
                    i-=1
                    temp="1"+curr_val
                    curr_val=temp
                i/=2
            if len(curr_val)==0:
                bin_vals.append("0")
            else:
                bin_vals.append(curr_val)
        max_len=len(bin_vals[0])
        for i in range(len(bin_vals)):
            if len(bin_vals[i])>max_len:
                max_len=len(bin_vals[i])

        for i in range(len(bin_vals)):
            bin_vals[i]=bin_vals[i].zfill(max_len)
        
        count_1s=[]
        for j in range(max_len):
            count=0
            for i in bin_vals:
                print(i)
                if int(i[j])==1:
                    count+=1
            count_1s.append(count)
        for i in range(max_len):
            if count_1s[i]%3==0:
                count_1s[i]=0
            else:
                count_1s[i]=1
        number=0
        for j in range(1,max_len+1):
            number+=2**(j-1)*count_1s[-j]
        return number
```
